### Research Question  (Is There a Correlation Between The Number Of Patient Who Showed and Each of  Gender, Age, Neighbourhood, Scholarship, Hipertension, Diabetes, Alcoholism,  Handcap, SMS_received)

```python
# First Assigning new names :
show =  df.No_show == 'No'
no_show = df.No_show == 'Yes'
```

```python
# show the number of patient who showed
df[show].count()
```

```python
# Show the number of patient who did not show
df[no_show].count()
```

From the previous: The number of patient who showed is about 4 times those who did not show.

$\frac{88208}{22319}$=4

```python
# plot between the number of patient who showed or not and gender
plt.figure(figsize=[10,10])
df.Gender[show].hist(alpha=0.5,color='blue', label='show')
df.Gender[no_show].hist(alpha=0.5,color='red',label= 'no_show')

locations= [0,1]
Gender =['Female','Male']
plt.xticks(locations,Gender)

plt.xlabel('Gender', fontsize=20)
plt.ylabel('Number Of Patients', fontsize=20);
plt.title('Who showed according to gender', fontsize=20)

plt.legend()
plt.show()
```

Gender is insignificant

```python
# plot between the number of patient who showed or not and Age
plt.figure(figsize=[10,10])
df.Age[show].hist(alpha=0.5,color='blue', label='show')
df.Age[no_show].hist(alpha=0.5,color='red', label= 'no_show')

plt.xlabel('Age', fontsize=20)
plt.ylabel('Number Of Patients', fontsize=20);
plt.title('who showed according to Age', fontsize=20)

plt.legend()
plt.show();
```

It is clear that the number of younger patient who showed is larger than the older

```python
# show the number of patient who showded according to the lacation of the hospital
plt.figure(figsize=[20,20])
df.Neighbourhood[show].value_counts().plot(kind='pie',autopct='%1.1f%%');
```

```python
df.Neighbourhood[show].value_counts()
```

It is clear that the location of the hospital affects the number of showed patient 

```python
# plot between the number of patient who showed or not and scholarship
plt.figure(figsize=[10,10])
df.Scholarship[show].hist(alpha=0.5,color='blue', label='show')
df.Scholarship[no_show].hist(alpha=0.5,color='red',label= 'no_show')

locations= [0,1]
state =['Not enrolled','Enrolled']
plt.xticks(locations,state)

plt.xlabel('Scholarship',fontsize=20)
plt.ylabel('Number Of Patients', fontsize=20)
plt.title('who showed according to scholarship',fontsize=20)

plt.legend()
plt.show();
```

scholarship is insignificant

```python
# plot between the number of patient who showed or not and hipertension
plt.figure(figsize=[10,10])
df.Hipertension[show].hist(alpha=0.5,color='blue', label='show')
df.Hipertension[no_show].hist(alpha=0.5,color='red', label= 'no_show')

locations= [0,1]
state =['Don\'t suffer','Suffer']
plt.xticks(locations,state)

plt.xlabel('hipertension', fontsize=20)
plt.ylabel('Number Of Patients', fontsize=20)
plt.title('who showed according to hipertension', fontsize=20)

plt.legend()
plt.show();
```
